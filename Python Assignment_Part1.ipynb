```python
import requests
from bs4 import BeautifulSoup
```

```python
base_url = "https://www.amazon.in/s?k=bags&crid=2M096C61O4MLT&qid=1653308124&sprefix=ba%2Caps%2C283&ref=sr_pg_"
```

```python
base_url
```

```python
def scrape_product_info(url):
    response = requests.get(url)
    soup = BeautifulSoup(response.content, "html.parser")
    
    products = []
    
    # Find all product containers on the page
    product_containers = soup.find_all("div", class_="s-result-item")
    
    for container in product_containers:
        try:
            # Extract product details
            product_url = container.find("a", class_="a-link-normal").get("href")
            product_name = container.find("span", class_="a-text-normal").get_text()
            
            rating_tag = container.find("span", class_="a-icon-alt")
            if rating_tag:
                rating = rating_tag.get_text().split()[0]
            else:
                rating = "Not available"
            
            product_data = {
                "Product URL": "https://www.amazon.in" + product_url,
                "Product Name": product_name,
                "Rating": rating,
            }
            
            products.append(product_data)
        except AttributeError:
            # Handle the case where some information is missing
            pass
    
    return products

# Scrape multiple pages
total_pages = 20
all_products = []

for page in range(1, total_pages + 1):
    page_url = base_url + str(page)
    products_on_page = scrape_product_info(page_url)
    all_products.extend(products_on_page)

# Print the scraped data
for product in all_products:
    print("Product Name:", product["Product Name"])
    print("Product URL:", product["Product URL"])
    print("Rating:", product["Rating"])
    print()
```
